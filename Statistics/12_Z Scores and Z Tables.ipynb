## Standard Normal Variate

Any Random Variabe Distribution can be converted to a Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import norm
```

```python
# Generate 1000 normal random intergers with specified mean and std
draw = norm.ppf(np.random.random(100000), loc=1000, scale=200).astype(int)
# loc is mean and scale is standard deviation
sns.histplot(draw, kde=True, bins=50)
# sns.distplot(draw)
```

```python
z = []
for i in draw:
    z.append((i-1000)/200)
# the formula for P(x) with mean = 1000, std = 200

a = np.array(z)
```

```python
print(a.mean())
print(np.round(a.mean()))
# note that the mean is 1
```

```python
print(a.std())
print(np.round(a.std()))
# note that the Standard Deviation is 1
```

```python
# now for the standardised array
sns.histplot(a, kde=True, bins=50)
```

## Calculating Z-score

Problem 1: imagine a group of 300 students took a test. Rohit scored 700 out of 1000. The average score was 500 and the standard deviation was 120. Assuming that test scores followed normal distribution find out how well Rohit perormed in comparision to his peers

```python
# Step 1 - Calculating the Z Score
# mean = 500 and std = 120
z = (700 - 500)/120
print(z)
```

```python
# Step 2 - Lookup the Z table (from ztable.com)
# this has the area under the curve for all the values of Z!!!
# z is 1.666666667 this is split as 1.6 + 0.0666667
# from the table we get the value = 0.9515
a = 0.9515
```

```python
# Step 3 - Multiply by 100
# we will get the number of people who have scored below Rohit
a*100
# therefore Rohit performed better than 95% of his peers
```
